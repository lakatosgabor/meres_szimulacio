```python
import numpy as np
import matplotlib.pyplot as plt

# Az adathalmaz
#x = darab
#y = merleg_db

def show_correction_function(x,y):

  mask = (x > 0.9) & (y > 1)
  x = x[mask]
  y = y[mask]

  # Lineáris regresszió
  coefficients = np.polyfit(x, y, 1)  # Lineáris egyenes illesztése (1. fokú polinom)
  slope, intercept = coefficients  # Az egyenes meredeksége és metszéspontja

  # Trendvonal készítése
  trendline = slope * x + intercept

  # R^2 érték kiszámítása
  # Korrelációs együttható kiszámítása
  correlation_matrix = np.corrcoef(x, y)
  correlation_coefficient = correlation_matrix[0, 1]
  r_squared = correlation_coefficient ** 2  # R^2 érték

  # Diagram
  plt.scatter(x, y, color='blue', label='Adatok')  # Pontok megjelenítése
  plt.plot(x, trendline, color='red', label=f'Trendvonal: y = {slope:.4f}x + {intercept:.4f}')  # Trendvonal
  plt.xlabel('X értékek')
  plt.ylabel('Y értékek')
  plt.title('Pont diagramm és lineáris trendvonal')
  plt.legend()
  plt.grid(True)
  plt.show()

  # Trendvonal egyenlete és R^2
  print(f'A trendvonal egyenlete: y = {slope:.4f}x + {intercept:.4f}')
  print(f'Az R^2 érték: {r_squared:.4f}')


def correction_function(x,y):
  mask = (x > 0.9) & (y > 1)
  x = x[mask]
  y = y[mask]

  coefficients = np.polyfit(x, y, 1)  # Lineáris egyenes illesztése (1. fokú polinom)
  slope, intercept = coefficients  # Az egyenes meredeksége és metszéspontja

  # Trendvonal készítése
  trendline = slope * x + intercept

  # R^2 érték kiszámítása
  # Korrelációs együttható kiszámítása
  correlation_matrix = np.corrcoef(x, y)
  correlation_coefficient = correlation_matrix[0, 1]
  r_squared = correlation_coefficient ** 2  # R^2 érték

  if (slope < 0):
    slope*=-1

  return 1-slope, intercept
```

```python
import numpy as np
import matplotlib.pyplot as plt

def show_error_function(x, y, felirat):
    # Lineáris regresszió
    coefficients = np.polyfit(x, y, 1)  # Lineáris egyenes illesztése (1. fokú polinom)
    slope, intercept = coefficients  # Az egyenes meredeksége és metszéspontja

    # Trendvonal készítése
    trendline = slope * x + intercept

    residuals = y - trendline

    # Kerekített érték vizsgálata
    mask = np.round(x) > 3  # Azok az indexek, ahol a kerekített x érték meghaladja a 92-t
    filtered_residuals = residuals[mask]

    # Maximális és minimális eltérés
    if len(filtered_residuals) > 0:
        max_deviation = np.max(np.abs(filtered_residuals))
        min_deviation = np.min(np.abs(filtered_residuals))
    else:
        max_deviation = min_deviation = None

    # R^2 érték kiszámítása
    correlation_matrix = np.corrcoef(x, y)
    correlation_coefficient = correlation_matrix[0, 1]
    r_squared = correlation_coefficient ** 2  # R^2 érték

    # Diagram
    plt.scatter(x, y, color='blue', label='Adatok')  # Pontok megjelenítése
    plt.plot(x, trendline, color='red', label=f'Trendvonal: y = {slope:.4f}x + {intercept:.4f}')  # Trendvonal
    plt.xlabel('X értékek')
    plt.ylabel('Y értékek')
    plt.title(felirat)
    plt.legend()
    plt.grid(True)
    plt.show()

    # Trendvonal egyenlete és R^2
    print(f'A trendvonal egyenlete: y = {slope:.4f}x + {intercept:.4f}')
    print(f'Az R^2 érték: {r_squared:.4f}')

    # Kerekített 92 fölötti pontok kilengései
    if len(filtered_residuals) > 0:
        print(f"Szórás: {np.std(filtered_residuals)}")
        print("Min eltérés: ", min_deviation, "Max eltérés: ", max_deviation)
    else:
        print("Nincs olyan adatpont, ahol a kerekített x érték meghaladná a 92-t.")
```

```python
import numpy as np
import pandas as pd
import random


def merleg_adatok(file_path, start_time, end_time):
    # Adatok beolvasása
    df = pd.read_csv(file_path, sep=';', header=None, names=['IDŐ', 'MERLEG_G'], parse_dates=['IDŐ'])

    # Szűrés az időintervallumra
    mask = (df['IDŐ'] >= start_time) & (df['IDŐ'] <= end_time)
    filtered_df = df.loc[mask]

    # Csak a második oszlop visszaadása listában
    return filtered_df['MERLEG_G'].sort_values(ascending=True).tolist()
```

```python
def meresek(file_path):
    df = pd.read_csv(file_path, sep=';', header=None)
    df = df.replace(',', '.', regex=True)
    df = df.astype(float)
    return df.values.flatten()  # 2D → 1D vektor
```
