## SZIMULÁCIÓ

```python
import numpy as np
import pandas as pd
import random

def merleg():
  # Paraméterek
  product_weight = 48.44  # termék tömege
  total_products = 112  # összes termék
  min_units = 1  # minimum termék egy rakásra
  max_units = 8  # maximum termék egy rakásra

  # Véletlenszerűen generálunk rakásokat, de figyelünk arra, hogy összesen 112 termék legyen
  rack_sizes = []
  remaining_products = total_products

  while remaining_products > 0:
      # Véletlenül választunk egy rakás méretet, de nem haladhatja meg a maradék termékek számát
      rack_size = np.random.randint(min_units, min(max_units, remaining_products) + 1)
      rack_sizes.append(rack_size)
      remaining_products -= rack_size

  # Az aktuális doboz súlyának számítása
  beerkezo_netto_suly_adatok = np.cumsum(np.array(rack_sizes) * (product_weight + random.uniform(-0.5, 0.5)))

  return beerkezo_netto_suly_adatok

merleg()
```

```python
for i in range (5):
  beerkezo_netto_suly_adatok = merleg()
  merleg_db = np.array([]) #A mérlegből beérkező darabszám, nettó tömeg / termék súly
  darab = np.array([]) #A mérlegből beérkezett darabszám kerekítve
  product_weight = 48.44

  for netto_suly in beerkezo_netto_suly_adatok:
    merleg_db = np.append(merleg_db, netto_suly / product_weight)
    darab = np.append(darab, np.round(netto_suly / product_weight))

  maradek = merleg_db - darab

  a,b = error_function(darab, maradek)

  correction_value = merleg_db - b + a * darab
  print(i, "--->", correction_value[-1], "--->", beerkezo_netto_suly_adatok[-1], beerkezo_netto_suly_adatok[-1] / product_weight)
```
